# Modal quickstart

Minimal example of defining a [Modal](https://modal.com) app and function, then running it both locally and remotely from a notebook.

```python
import sys
import modal

app = modal.App("example-hello")
```

A `modal.App` is the top-level container that groups related functions, images, and volumes together. It's referenced by name (`example-hello`) both locally and on the Modal dashboard.

```python
@app.function()
def f(message):
    print(f"modal message {message}")
```

The `@app.function()` decorator registers `f` with the app so it can be scheduled to run in a remote Modal container. It's still an ordinary Python function otherwise — Modal serializes it with cloudpickle, so it works fine even though it's defined directly in a notebook cell.

```python
f.local("hello")
```

`.local(...)` just calls the function in-process, like calling `f(...)` directly — no Modal infrastructure involved. Useful for testing the function body without touching the network.

`.remote(...)` runs the function on Modal's infrastructure instead of locally. Calling it directly fails with an "app is not running" error, because the function needs to be "hydrated" with metadata (container image, app id, etc.) that only exists while the app is actively running.

When you run a script with `modal run script.py`, the CLI wraps execution in an app context automatically. In a notebook there's no such wrapper, so you need to open that context yourself with `with app.run():`.

`modal.enable_output()` additionally streams the remote container's stdout/stderr back into the notebook — otherwise `.remote()` still works, you just wouldn't see the `modal message hello` print until the call returns. The `View run at https://modal.com/apps/...` link in the output below takes you to the same logs and container details on the Modal dashboard.

```python
with modal.enable_output():
    with app.run():
        f.remote("hello")
```
